````python
## 基础模型推理结果提取
import json
            

ori_query = [
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/gemma9B_base_normal_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/gemma9B_base_schema_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/gemma9B_base_schema_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/gemma9B_base_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/LLama8B_base_normal_prompt.json",
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/LLama8B_base_schema_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/LLama8B_base_schema_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/LLama8B_base_vector_prompt.json",
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/Qwen2.57B_base_normal_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/Qwen2.57B_base_schema_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/Qwen2.57B_base_schema_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/Qwen2.57B_base_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/GLM49B_base_normal_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/GLM49B_base_schema_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/GLM49B_base_schema_vector_prompt.json", 
  "/home/work/liuytest/LLaMA-Factory/src/llamafactory/chat/GLM49B_base_vector_prompt.json", 
]


import re 
pattern = r'```(cypher)?\n(.*?)\n```'
pattern_label = r'\[[\d\.]+\]'

for query_file in ori_query:
    print(query_file)
    all = query_file.split("/")
    file_name = all[len(all)-1]
    result = []
    
    query_pair = None
    with open(query_file, 'r',encoding="utf-8") as json_file:
        key_values = json.load(json_file)
    
    for query_row in key_values:
        try:
            output = query_row["output"]
            # 使用 re.search 查找匹配项
            match = re.search(pattern, output, re.DOTALL)  
             # 提取捕获组中的内容
            cypher = match.group(2).strip()
            
            # 使用re.sub函数进行替换
            cypher = re.sub(pattern_label, '', cypher)

            result.append({"instruction":query_row["instruction"],"output":cypher})
        except Exception as e:
            print("errrrrrrrrrrrrrrr",query_file,e,output)
            result.append({"instruction":query_row["instruction"],"output":None})
    # 方法一：使用json.dump()函数保存为JSON格式并写入文件
    with open(f"不规则输出Cypher提取/{file_name}_fix.json", "w") as file:
        json.dump(result, file,indent=4,ensure_ascii=False,)
        
````
