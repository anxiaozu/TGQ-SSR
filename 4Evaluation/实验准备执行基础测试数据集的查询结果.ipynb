```python
## 通过数据库的base_schema pattern 查询数据库 生成带有 where return order by
import json
import psycopg

ori_query = [
    "/home/work/liuytest/大论文/2微调/ldbc_normal_test_7_3.json", 
                                                                                     ######## 此处有问题 会被覆盖 请单独处理
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_1_9.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_2_8.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_3_7.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_4_6.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_5_5.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_6_4.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_7_3.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_8_2.json",
    "/home/work/liuytest/大论文/2微调/scala_with_schema_vector_prompt_test_9_1.json"
]

pg_con = psycopg.connect(
            dbname="postgres",
            host="localhost",
            user="omm",
            password="ZAQ1234@esz",
            port=5432,
            options="-c statement_timeout=10000"
            ) 
pg_con.autocommit=True        
cur = pg_con.cursor()
cur.execute("set graph_path = ldbc")


for query_file in ori_query:
    print(query_file)
    all = query_file.split("/")
    file_name = all[len(all)-1]
    result = []
    
    query_pair = None
    with open(query_file, 'r',encoding="utf-8") as json_file:
        key_values = json.load(json_file)
    
    for query_row in key_values:
        try:    
            cur.execute(query_row["output"])
            # 这里应该把结果处理成表格  最后就好计算了
            rows = cur.fetchall()
            column_names = [desc[0] for desc in cur.description]
            print("数据的总行数:",len(rows))
            print("数据的总列数:",len(column_names))
            query_result = len(rows)
            result.append({"instruction":query_row["instruction"],"output":query_row["output"],"column_names":column_names,"rows":rows,"err":False})
        except Exception as e:
            result.append({"instruction":query_row["instruction"],"output":None,"column_names":None,"rows":None,"err":True})    
    # 方法一：使用json.dump()函数保存为JSON格式并写入文件
    with open(f"{file_name}_result.json", "w") as file:
        json.dump(result, file,indent=4,ensure_ascii=False,)
        
pg_con.close()    
```
